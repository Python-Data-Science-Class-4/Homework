```python
import psycopg2

def terminate_database_connections(database_name):
    try:
        connection = psycopg2.connect(
            "dbname=postgres user=postgres password=Ysf55Ayd08 host=localhost port=5432"
        )
        connection.autocommit = True

        # Get the PID of connections to the database
        with connection.cursor() as cursor:
            cursor.execute(
                f"SELECT pg_terminate_backend(pg_stat_activity.pid) "
                f"FROM pg_stat_activity "
                f"WHERE pg_stat_activity.datname = '{database_name}' AND pid <> pg_backend_pid();"
            )

        print(f"Connections to the '{database_name}' database terminated successfully.")
    except psycopg2.Error as e:
        print(f"Error terminating connections: {e}")
    finally:
        if connection:
            connection.close()

# Terminate connections before dropping the database
terminate_database_connections("class4")

# Drop the database
try:
    conn = psycopg2.connect("dbname=postgres user=postgres password=Ysf55Ayd08 host=localhost port=5432")
    conn.autocommit = True
    command = "DROP DATABASE IF EXISTS class4;"
    cur = conn.cursor()
    cur.execute(command)
    cur.close()
except psycopg2.Error as e:
    print(f"Error dropping database: {e}")
finally:
    if conn:
        conn.close()
```

FROM film
JOIN inventory ON film.film_id = inventory.film_id
LEFT JOIN rental ON inventory.inventory_id = rental.inventory_id
GROUP BY film.title
ORDER BY rental_count ASC
LIMIT 5;
