# 2- Categorization

## Categorize satisfaction in 3 level and label them as 'Low', 'Moderate' and 'High'. Add a new column named 'Satisfaction Category' and store them there

#### ||| Low Satisfaction : (1-3) ||| Moderate Satisfaction : (4-7) ||| High Satisfaction : (8-10) |||

```python
mask1=data_full['Satisfaction Level'].between(1,3)
mask2=data_full['Satisfaction Level'].between(4,7)
mask3=data_full['Satisfaction Level'].between(8,10)

data_full.loc[mask1, 'Satisfaction Category']='Low Satisfaction'
data_full.loc[mask2, 'Satisfaction Category']='Moderate Satisfaction'
data_full.loc[mask3, 'Satisfaction Category']='High Satisfaction'

data_full

```

## Categorize income in 3 level and label them as 'Low', 'Moderate' and 'High'. Add a new column named 'Income Category' and store them there

```python
maks=data_full['Income'].max()
min=data_full['Income'].min()

print(maks)
print(min)

mask1=data_full['Income'].between(0,25000)
mask2=data_full['Income'].between(25001,50000)
mask3=data_full['Income'].between(50001,75000)

data_full.loc[mask1, 'Income Category']='Low'
data_full.loc[mask2, 'Income Category']='Moderate'
data_full.loc[mask3, 'Income Category']='High'

data_full
```
