```python
#Rock Paper Scissor
import random

class Player():
    def __init__(self,Player,pscore=0) -> None:
        self.player=Player

    def pmove(self):
        return int(input("1:Stone,2:Paper,3:Scissors"))

class Computer():
    def __init__(self,cscore=0) -> None:
        pass

    def cmove(self):
        return random.randint(1,3)

class Game(Player,Computer):
    def __init__(self,Player,pscore=0) -> None:
        super().__init__(Player,pscore=0)
        self.round=0
        self.pscore=0
        self.cscore=0
        while True:
            if self.round>0:
                if input("one more round? (Y/N)").lower()=="n":
                    break
            self.round+=1
            s1=self.pmove()
            s2=self.cmove()
            if s1!=s2:
                if s1==1 and s2==3:
                    self.pscore+=1
                    print(f"Round: {self.round}, {self.player} wins this round")
                    print(f'''
Player Score   : {self.pscore}
Computer Score : {self.cscore}
                          ''')
                elif s1==2 and s2==2:
                    self.pscore+=1
                    print(f"Round: {self.round}, {self.player} wins this round")
                    print(f'''
Player Score   : {self.pscore}
Computer Score : {self.cscore}
                          ''')
                elif s1==2 and s2==1:
                    self.pscore+=1
                    print(f"Round: {self.round}, {self.player} wins this round")
                    print(f'''
Player Score   : {self.pscore}
Computer Score : {self.cscore}
                          ''')
                elif s1==2 and s2==3:
                    self.cscore+=1
                    print(f"Round: {self.round}, Computer wins this round")
                    print(f'''
Player Score   : {self.pscore}
Computer Score : {self.cscore}
                          ''')
                elif s1==3 and s2==1:
                    self.cscore+=1
                    print(f"Round: {self.round}, Computer wins this round")
                    print(f'''
Player Score   : {self.pscore}
Computer Score : {self.cscore}
                          ''')
                elif s1==3 and s2==2:
                    self.cscore+=1
                    print(f"Round: {self.round}, Computer wins this round")
                    print(f'''
Player Score   : {self.pscore}
Computer Score : {self.cscore}
                          ''')
            else:
                print(f'''
Round: {self.round}, This round is a draw
{self.player}: {s1}
Computer : {s2}
''')

        if self.pscore>self.cscore:
            print(self.player+" won")
        elif self.pscore<self.cscore:
            print("Computer won")
        else:
            print("It's a draw")        
```
